```python
import pandas as pd
from tqdm import tqdm
import json
import ast
import os
```

```python
dir_path = os.getcwd()
print(dir_path)
```

```python
csv_file_path = './data/CONAN/CONANSP.csv'
txt_file_path = './data/CONANSP_for_VAE.txt'
json_file_path = './data/CONAN/CONANSP.json'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path, encoding='latin1')
column_data = df['counterSpeech']

with open(txt_file_path, 'w', encoding='utf-8') as file:
    for value in column_data:
        file.write(str(value) + '\n')
```

```python
# Define the JSON structure
data = {"conan": []}

# Process each row in the DataFrame and add it to the JSON structure
for _, row in df.iterrows():
    entry = {
        "cn_id": row["cn_id"],
        "hateSpeech": row["hateSpeech"],
        "counterSpeech": row["counterSpeech"],
        "hsType": row["hsType"],
        "hsSubType": row["hsSubType"],
        "cnType": row["cnType"],
        "age": row["age"],
        "gender": row["gender"],
        "educationLevel": row["educationLevel"]
    }
    data["conan"].append(entry)

# Write the JSON structure to a file with UTF-8 encoding
with open(json_file_path, 'w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=4)

print(f"CSV file has been successfully converted to JSON and saved at {json_file_path}")
```
